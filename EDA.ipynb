```python
# Step 1: Import Pandas
import pandas as pd

# Step 2: Load the dataset

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Step 3: Display the dataset shape (rows, columns)
print("Dataset Shape (rows, columns):")
print(df.shape)

# Step 4: Preview the first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Step 5: Check column names
print("\nColumn Names:")
print(df.columns)

# Step 6: Check data types and missing values
print("\nDataset Information:")
print(df.info())
```

```python
import matplotlib.pyplot as plt

df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count in Titanic Dataset")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a new column: Group
def passenger_group(row):
    if row["Age"] < 16:
        return "Child"
    elif row["Sex"] == "female":
        return "Woman"
    else:
        return "Man"

df["Group"] = df.apply(passenger_group, axis=1)

# Survival rate by group
survival_rates = df.groupby("Group")["Survived"].mean()

print(survival_rates)

# Plot
survival_rates.plot(kind="bar")
plt.title("Survival Rate: Men vs Women vs Children")
plt.ylabel("Survival Probability")
plt.show()
```
